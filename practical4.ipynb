```python
import geemap.geemap as map
```

```python
import ee
```

```python
import os
```

```python
ee.Authenticate()
```

```python
ee.Initialize(project = 'shreenitya0428')
```

```python
m = map.Map()
m
```

```python
Indian_districts = ee.FeatureCollection('projects/shreenitya0428/assets/India_districts')
m.addLayer(Indian_districts,{},'Indian_districts')
m
```

```python
Treecover_2019 = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2019-01-01","2019-12-31").select("Percent_Tree_Cover").mean().clip(Indian_districts)
Visualize_Tree_2019 = {"min":0, "max":100, 'palette': ['bbe029', '0a9501', '074b03']}
m.addLayer(Treecover_2019, Visualize_Tree_2019, 'Treecover_2019')
m
```

```python
out_dir = os.path.join(os.path.expanduser('~'), 'Downloads')
India_treecover_2019 = os.path.join(out_dir, 'India_treecover_2019.csv')
if not os.path.exists(out_dir):
    os.makedirs(out_dir)
map.zonal_stats(Treecover_2019, Indian_districts, India_treecover_2019, statistics_type='MEAN', scale=250)
```
